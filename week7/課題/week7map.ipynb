```python
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
hiroshima_ku = gpd.read_file('N03-23_34_230101.geojson')
```

```python
# とりあえず変数に入れる
hiroshima_ku.plot()
```

```python
# とりあえず変数に入れる
hiroshima_ku.plot
```

```python
# 最初の５列は？
hiroshima_ku.head()
```

```python
# データの情報
hiroshima_ku.info()
```

```python
hiroshima_pop = pd.read_csv('hiroshima.csv')
```

```python
hiroshima_pop.head()
```

```python
hiroshima_pop.info()
```

```python
# rename
hiroshima_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
# check the new fieldname
hiroshima_ku
```

```python
# OBJECTからintに変える
hiroshima_ku['KEY_CODE'] = hiroshima_ku['KEY_CODE'].astype(int)
```

```python
# checkする
hiroshima_ku.info
```

```python
# ここでようやく merge コマンドで統合させて、新しい変数に入れる
hiroshima_ku_pop = hiroshima_ku.merge(hiroshima_pop,on='KEY_CODE')
```

```python
# check!
hiroshima_ku_pop
```

```python
hiroshima_ku_pop.info()
```

```python
# change from OBJECT to int
hiroshima_ku_pop['人口総数'] = hiroshima_ku_pop['人口総数'].astype(int)
```

```python
# check
hiroshima_ku_pop.info()
```

```python
hiroshima_ku_pop.plot(column='人口総数',legend=True)
```
